```python
options = {
    "sort":"relevance",
    "news_desk":["Business Day",
                 "Business",
                 "DealBook",
                 "Financial",
                 "Market Place",
                 "Technology",
                 "U.S."],
    "type_of_material": ["News Analysis", 
                         "News", 
                         "Article"]}
@sleep_and_retry
@limits(calls=10, period=60, raise_on_limit = True)
def fetch_articles(company):
    
    try:
        response = nyt.article_search(query=company, results=1, dates={'begin':start_date, 'end':end_date}, options=options)
        response = [d for d in response if isinstance(d, dict)]
        if company not in article_dict:
            
            article_dict[company] = response
        
        else: 
            
            article_dict[company].append(response)
            
        return article_dict
    
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
company_list = ['Apple Inc.', 'Adobe Inc.']
article_dict = {}
test_df = pd.DataFrame()
```

```python
def article_dic_to_df (test_df, company):
    temp_df = pd.DataFrame.from_records(article_dict[company])
    temp_df.columns = pd.MultiIndex.from_product([[company], temp_df.columns])
    return pd.concat([test_df, temp_df], axis = 1)
```

```python
for company in company_list:
    fetch_articles(company)
    test_df = article_dic_to_df (test_df, company)
```
